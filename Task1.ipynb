```python
# Summary of the data (ony numerical)
df.describe()
```

```python
# scatterplot
sns.scatterplot(x='sepal_length', y='sepal_width', data=df ,hue = "species")
```

***It shows the relationship between `sepal length `and `sepal width`, with different colors representing the three `species` of iris flowers.***
- ***`Setosa` (Blue dots): This species tends to have smaller `sepal length` and larger `sepal width` compared to the other two species. They form a distinct cluster at the bottom-left of the plot.***
- ***`Versicolor` (Orange dots): This species generally falls in the middle range for both `sepal length` and `sepal width`, showing some overlap with both Setosa and Virginica.***
- ***`Virginica` (Green dots): This species typically has larger `sepal length` and a wider range of `sepal width`, often clustering towards the upper-right part of the plot.***

```python
# scatterplot
sns.scatterplot(x='petal_length', y='petal_width', data=df ,hue = "species")
```

 ***It shows the relationship between `petal length` and `petal width`, with different colors again representing the three `species` of iris flowers.***
- ***`Setosa` (Blue dots): This species is very clearly separated from the other two. They exhibit the smallest `petal length` and `petal width`, forming a tight, distinct cluster at the bottom-left of the plot.***
- ***`Versicolor` (Orange dots): This species generally falls in the middle range for both `petal length` and `petal width`. Its cluster is located between Setosa and Virginica, showing some overlap with Virginica but still largely distinct from Setosa.***
- ***`Virginica` (Green dots): This species typically has the largest `petal length` and `petal width`, clustering towards the upper-right part of the plot. There's some overlap with Versicolor, but Virginica generally occupies the region of longer and wider petals.***

```python
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
sns.histplot(df['sepal_length'], kde=True, bins=10)
plt.title('Distribution of Sepal Length')

plt.subplot(2, 2, 2)
sns.histplot(df['sepal_width'], kde=True, bins=10)
plt.title('Distribution of Sepal Width')

plt.subplot(2, 2, 3)
sns.histplot(df['petal_length'], kde=True, bins=10)
plt.title('Distribution of Petal Length')

plt.subplot(2, 2, 4)
sns.histplot(df['petal_width'], kde=True, bins=10)
plt.title('Distribution of Petal Width')

plt.tight_layout()
plt.show()
```

 ***So, the interpretation of the above charts are :***
- ***Distribution of `Sepal Length`: This histogram appears to be somewhat normally distributed, with the majority of sepal lengths falling between 5.0 cm and 7.0 cm.***
- ***Distribution of `Sepal Width`: This distribution looks roughly bell-shaped, centered around 3.0 cm to 3.5 cm, with a slight positive skew. There's a clear peak around these values.***
- ***Distribution of `Petal Length`: This histogram shows a distinctly bimodal or even trimodal distribution. There's a strong peak at very short petal lengths (around 1.0 cm to 2.0 cm), corresponding to the 'setosa' species, and then a broader distribution for longer petal lengths (around 4.0 cm to 6.0 cm), likely representing 'versicolor' and 'virginica' species. This suggests petal length is a strong differentiating feature.***
- ***Distribution of `Petal Width`: Similar to petal length, the distribution of petal width also appears to be multimodal, with a sharp peak at smaller widths (around 0.2 cm) and another, broader distribution for larger widths. This further indicates that petal width is also a very effective feature for distinguishing between the iris species.***

```python
# check outliers in data
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
sns.boxplot(y=df['sepal_length'])
plt.title('Box Plot of Sepal Length')

plt.subplot(2, 2, 2)
sns.boxplot(y=df['sepal_width'])
plt.title('Box Plot of Sepal Width')

plt.subplot(2, 2, 3)
sns.boxplot(y=df['petal_length'])
plt.title('Box Plot of Petal Length')

plt.subplot(2, 2, 4)
sns.boxplot(y=df['petal_width'])
plt.title('Box Plot of Petal Width')

plt.tight_layout()
plt.show()
```

- ***Box Plot of `Sepal Length`: This plot shows that the sepal length distribution is quite uniform, with no significant outliers. All data points fall within the whiskers, indicating a consistent range.***
- ***Box Plot of `Sepal Width`: This plot indicates the presence of a few outliers. There are data points both above the upper whisker (values greater than approximately 4.0 cm) and below the lower whisker (values less than approximately 2.2 cm). These represent sepal widths that are unusually large or small compared to the majority of the data.***
- ***Box Plot of `Petal Length`: Similar to sepal length, the petal length distribution does not show any obvious outliers. The data points are contained within the whiskers.***
- ***Box Plot of `Petal Width`: The petal width distribution also appears to be without significant outliers. All values are within the typical range defined by the box and whiskers.***
